```python
import pandas as pd
```

```python
train = pd.read_csv('dataset/train.csv', index_col='id')
train.head()
```

```python
test = pd.read_csv('dataset/test.csv', index_col='id' )
test.head()
```

```python
train.info()
```

```python
word_to_num = {"white":0, "red":1}

train['type'] = train['type'].replace(word_to_num)
train
```

```python
test['type'] = test['type'].replace(word_to_num)
test
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
s_col = list(train.columns)
s_col.pop(-1)
s_col.pop(-1)
s_col
```

```python
train[s_col].shape
```

```python
m_scaler = MinMaxScaler()

train_mscale = m_scaler.fit_transform(train[s_col])
df_train = pd.DataFrame(train_mscale, columns=s_col)
df_train.head()
```

```python
df_train.columns
f_names = ['fixed acidity', 
           'volatile acidity', 'citric acid', 
           'residual sugar',
           'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 
           'sulphates',
           'alcohol']
f_names
```

```python
X_train = df_train[f_names]
X_train
```

```python
y_train = train['quality']
y_train
```

```python
test_mscale = m_scaler.fit_transform(test[s_col])
df_test = pd.DataFrame(test_mscale, columns=s_col)
df_test
```

```python
X_test = df_test[f_names]
X_test
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = { 'n_estimators' : [10, 30, 50, 100, 150, 200, 300],
           'min_samples_split' : [3, 5, 10, 15],
          'min_samples_leaf' : [2, 5,10],
          'max_depth' : [6, 8, 10, 12],
            }
```

```python
rf_clf = RandomForestClassifier(random_state = 0, n_jobs = -1)
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 3, n_jobs = -1)
grid_cv.fit(X_train, y_train)
```

```python
print('<최적 하이퍼 파라미터>')
print(grid_cv.best_params_)
```

```python
new_rf_clf = RandomForestClassifier(n_estimators = 150, 
                                min_samples_leaf = 2,
                                min_samples_split = 3,
                                max_depth = 12)
new_rf_clf.fit(X_train, y_train)
pred = new_rf_clf.predict(X_test)
pred
```

```python
import numpy as np
```

```python
def ACCURACY(true, pred):   
    score = np.mean(true==pred)
    return score
```

```python
score = ACCURACY(y_train, pred)

print(f"모델의 정확도는 {score*100:.2f}% 입니다")
```
